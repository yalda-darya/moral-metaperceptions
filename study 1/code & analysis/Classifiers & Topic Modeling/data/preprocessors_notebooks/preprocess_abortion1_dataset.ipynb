```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
```

```python
abortion1_df = pd.read_csv(os.path.join(os.path.dirname(os.getcwd()), "raw", "abortion_sentiment_prediction", "abortion1.csv"))
abortion1_df
```

```python
# Value count for each of the data_source
abortion1_df["data_source"].value_counts()
```

```python
# lower case everything in "annotation" column
abortion1_df["annotation"] = abortion1_df["annotation"].str.lower()

# strip everything in "annotation" column
abortion1_df["annotation"] = abortion1_df["annotation"].str.strip()

# Get the unique values in "annotation" column
abortion1_df["annotation"].unique()
```

```python
# throw out the nan rows
abortion1_df = abortion1_df[~abortion1_df["annotation"].isna()]
abortion1_df["annotation"].unique()
```

```python
abortion1_df["annotation"].value_counts()
```

```python
abortion1_df.loc[:, "annotation"] = abortion1_df.loc[:, "annotation"].replace(to_replace=["neutral", "throw out"], value="throw out")
abortion1_df["annotation"].unique()
```
