```python
import pandas as pd
import os
```

```python
# Get parent directory
parent_dir = os.path.dirname(os.getcwd())
```

```python
mortezas_dataset_complete_with_annotations_path = os.path.join(
    parent_dir, "raw", "mortezas_dataset_complete_with_annotations.csv"
)
mortezas_dataset_complete_with_annotations_path
```

```python
isi_dataset_sampled_with_annotations_path = os.path.join(
    parent_dir, "raw", "isi_dataset_sampled_with_annotations.csv"
)
isi_dataset_sampled_with_annotations_path
```

```python
mortezas_dataset_complete_with_annotations_df = pd.read_csv(
    mortezas_dataset_complete_with_annotations_path,
    engine="python",
    dtype={
        "full_text": "str",
        "conversation_id_str": "str",
        "created_at": "object",
        "id_str": "str",
        "user": "object",
        "annotation": "str",
    },
)
mortezas_dataset_complete_with_annotations_df
```

```python
# Add a column to mortezas_dataset_complete_with_annotations_df named "source" with the value "mortezas"
mortezas_dataset_complete_with_annotations_df["source"] = "mortezas"
mortezas_dataset_complete_with_annotations_df
```

```python
# Check for nulls in mortezas_dataset_complete_with_annotations_df
mortezas_dataset_complete_with_annotations_df.isnull().sum()
```

```python
isi_dataset_sampled_with_annotations_df = pd.read_csv(
    isi_dataset_sampled_with_annotations_path,
    engine="python",
    dtype={
        "full_text": "str",
        "conversation_id_str": "str",
        "created_at": "object",
        "id_str": "str",
        "user_id_str": "str",
        "annotation": "str",
    },
)
isi_dataset_sampled_with_annotations_df
```

```python
# Add a column to isi_dataset_sampled_with_annotations_df named "source" with the value "isi"
isi_dataset_sampled_with_annotations_df["source"] = "isi"
isi_dataset_sampled_with_annotations_df
```

```python
# Rename the column "user_id_str" to "user" in isi_dataset_sampled_with_annotations_df
isi_dataset_sampled_with_annotations_df.rename(
    columns={"user_id_str": "user"}, inplace=True
)
isi_dataset_sampled_with_annotations_df
```

```python
# Make sure the columns in mortezas_dataset_complete_with_annotations_df and isi_dataset_sampled_with_annotations_df are the same
assert mortezas_dataset_complete_with_annotations_df.columns.tolist() == isi_dataset_sampled_with_annotations_df.columns.tolist()
```

```python
# Concatenate the two dataframes
df = pd.concat(
    [mortezas_dataset_complete_with_annotations_df, isi_dataset_sampled_with_annotations_df]
)
df
```

```python
# Are there null values in the dataframe?
df.isnull().sum()
```

```python
df.dtypes
```

```python
# Get annotation unique values
df["annotation"].unique()
```

```python
# Strip the annotation column
df["annotation"] = df["annotation"].str.strip()
df["annotation"].unique()
```

```python
# lower case the annotation column
df["annotation"] = df["annotation"].str.lower()
df["annotation"].unique()
```

```python
# Change every "throw out" label to "throw_out" under the "annotation" column
df["annotation"] = df["annotation"].str.replace("throw out", "throw_out")
df["annotation"].unique()
```

```python
df["annotation"].value_counts()
```

```python
# Drop rows with "unsure" value in their "annotation" column
df = df[df["annotation"] != "unsure"]
df["annotation"].value_counts()
```

```python
# reset the index
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv(os.path.join(parent_dir, "raw", "abortion_dataset_complete.csv"), index=False)
```
