```python
import pandas as pd
import os
```

# Abortion1 Dataset

```python
original_file = pd.read_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'reproduction_tweets.csv'), usecols=['id_str', 'created_at', 'full_text', 'lang', 'user.id', 'user.lang', 'user.location', 'created_at_date', 'geo.coordinates[0]', 'geo.coordinates[1]', 'geo.type'])
original_file
```

```python
original_file['id_str'] = original_file['id_str'].astype(str)
original_file['created_at'] = original_file['created_at'].astype(str)
original_file['full_text'] = original_file['full_text'].astype(str)
original_file['lang'] = original_file['lang'].astype(str)
```

```python
predicted_file = pd.read_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'reproduction_tweets_with_predictions.csv'), usecols=['full_text', 'id_str', 'created_at', 'annotation'])
predicted_file
```

```python
predicted_file['id_str'] = predicted_file['id_str'].astype(str)
predicted_file['full_text'] = predicted_file['full_text'].astype(str)
predicted_file['created_at'] = predicted_file['created_at'].astype(str)
predicted_file['annotation'] = predicted_file['annotation'].astype(str)
predicted_file
```

```python
assert len(original_file) == len(predicted_file)
```

```python
# These two files have the same number of rows, and the id_str column is the same in both files.
# I want to merge these two, keeping all of the original's columns, **only** the annotation column from the predicted file.
# I want to do this by matching the id_str column.
merged_df = pd.merge(original_file, predicted_file[['id_str', 'annotation']], on='id_str', how='left')
merged_df
```

```python
merged_df['annotation'].value_counts()
```

```python
merged_df.to_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'reproduction_tweets_with_predictions.csv'), index=False)
```

```python
original_file = pd.read_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'all_geotagged_sentiment.csv'))
original_file
```

```python
original_file['conversation_id'] = original_file['conversation_id'].astype(str)
original_file['created_at'] = original_file['created_at'].astype(str)
original_file['text'] = original_file['text'].astype(str)
original_file
```

```python
predicted_file = pd.read_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'all_geotagged_sentiment_with_predictions.csv'), usecols=['full_text', 'id_str', 'created_at', 'annotation'])
predicted_file
```

```python
predicted_file['id_str'] = predicted_file['id_str'].astype(str)
predicted_file['full_text'] = predicted_file['full_text'].astype(str)
predicted_file['created_at'] = predicted_file['created_at'].astype(str)
predicted_file['annotation'] = predicted_file['annotation'].astype(str)
predicted_file
```

```python
# Drop the duplicate rows that have same id_str
predicted_file = predicted_file.drop_duplicates(subset=['id_str'])
predicted_file
```

```python
# the id_str column in predicted_df is the same as the conversation_id column in original_df
# I want to merge these two, keeping all of the original's columns, and **only** the annotation column from the predicted file.
merged_df = pd.merge(original_file, predicted_file[['id_str', 'annotation']], left_on='conversation_id', right_on='id_str', how='left')
merged_df
```

```python
# Drop the rows that have NaN in the annotation column
merged_df = merged_df.dropna(subset=['annotation'])
merged_df
```

```python
# reset the index
merged_df = merged_df.reset_index(drop=True)
merged_df
```

```python
# save the merged_df to a csv file
merged_df.to_csv(os.path.join(os.getcwd(), 'raw', 'abortion_sentiment_prediction', 'all_geotagged_sentiment_with_predictions.csv'), index=False)
```
